```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as LR
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import numpy as np
import joblib as jb

# Load the dataset
myData = pd.read_csv('/road_accident_severity.csv')

# Reshape x to be a 2D array before splitting
x = myData[['Total Accidents']]
y = myData['Total Deaths'].values.reshape(-1, 1)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)

# Initialize and train the model
model = LR()
model.fit(x_train, y_train)

# Make predictions on the test set
y_pred = model.predict(x_test)

# Calculate performance metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Plotting
plt.scatter(myData['Total Accidents'], myData['Total Deaths'], color='blue', label='Data Points')

# Predict on a range of values for plotting the regression line
# Ensure X_line is a 2D array with the correct shape
X_line = np.linspace(myData['Total Accidents'].min(), myData['Total Accidents'].max(), 100).reshape(-1, 1)
y_line = model.predict(X_line)

# Plot the regression line
plt.plot(X_line, y_line, color='red', label='Regression Line')

# Label the axes and show the legend
plt.xlabel('Total Accidents')
plt.ylabel('Total Deaths')
plt.title('Total Accidents vs Total Deaths')
plt.legend()
plt.grid(True)
plt.show()
print("Mean Squared Error:", mse)
print("R-squared:", r2)
jb.dump(model,"road_accident_severity.plk")
```
